```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
frames = []

cap = cv.VideoCapture('test_images/wmv.wmv')
success,image = cap.read()
success = True    
while success:
    success,frame = cap.read()
    if success:
        frame = frame.astype(np.uint8)
        frame = frame[50:550, 50:550]
        frame = cv.cvtColor(frame, cv.COLOR_RGB2BGR)
        frames.append(frame)

img = np.stack(frames, axis=0)
img = np.flip(img)
```
